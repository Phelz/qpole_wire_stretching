```python
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import data
```

```python
pTrace = pd.read_csv(data.DIR_PATH / 'particle_data' / 'particle_1.csv')
```

```python
def trim_ptrace(pTrace):
    
    # Only keep x, y and z columns
    pTrace = pTrace[['x', 'y', 'z']]

    # Keep only the first of the duplicates and remove everything else from the original dataframe
    pTrace = pTrace.drop_duplicates(subset=['x', 'y', 'z'], keep='first')
    
    return pTrace

# Save the new particle trace in a new csv file in data.DIR_PATH / 'particle_data_trimmed'
pTrace = trim_ptrace(pTrace)
pTrace.to_csv(data.DIR_PATH / 'particle_data_trimmed' / 'particle_1.csv', index=False)

# Go through all the unfrozen ptraces in data.DIR_PATH / 'particle_data' and trim them
for pTracePath in list(data.DIR_PATH.glob('particle_data/*.csv')):
    if 'frozen' in pTracePath.stem:
        print(f'Skipping {pTracePath.stem}')
        continue
    
    pTrace = pd.read_csv(pTracePath)
    pTrace = trim_ptrace(pTrace)
    pTrace.to_csv(data.DIR_PATH / 'particle_data_trimmed' / pTracePath.name, index=False)
```
